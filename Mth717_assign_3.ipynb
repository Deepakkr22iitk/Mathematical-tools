```python
def activation(inputs, weights, threshold):
    total = sum(x * w for x, w in zip(inputs, weights))
    if total >= threshold:
        return 1
    else:
        return 0

def run_logic_gates():
    # Standard binary input combinations
    test_inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]

    # Configuration for AND Gate
    # Threshold is high so both inputs must be 1 to trigger
    and_weights = [1, 1]
    and_threshold = 1.5

    # Configuration for OR Gate
    # Threshold is low so any input of 1 triggers the output
    or_weights = [1, 1]
    or_threshold = 0.5

    print("--- AND Gate Results ---")
    for val in test_inputs:
        res = activation(val, and_weights, and_threshold)
        print(f"Input: {val} | Output: {res}")

    print("\n--- OR Gate Results ---")
    for val in test_inputs:
        res = activation(val, or_weights, or_threshold)
        print(f"Input: {val} | Output: {res}")

if __name__ == "__main__":
    run_logic_gates()
```
